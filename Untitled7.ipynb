```python
import random
def estimatepi(number):
	Pointincircle = 0
	Totalpoints = 0

	for _ in range(number):
		X = random.uniform(0,1)
		Y = random.uniform(0,1)
		distance = X**2 + Y**2
		if distance <= 1:
			Pointincircle += 1
		Totalpoints += 1

	
	_pi = (Pointincircle * 4)/ Totalpoints

	return _pi

estimatepi(1000000000)
```
